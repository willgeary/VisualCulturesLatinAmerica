## Exploring the Google Street View API

API stands for *Application Programming Interface*. APIs enable us to get data programatically by making requests in a pre-defined format to a URL.

Google's [Street View Image API](https://developers.google.com/maps/documentation/streetview/) enables us to download images taken by Google's global army of Street View cars.

For example, check out: https://maps.googleapis.com/maps/api/streetview?size=800x800&location=40.8084322,-73.9649473&heading=120&pitch=0

Here's what that returns:

```python
from IPython.display import Image, HTML
print "https://maps.googleapis.com/maps/api/streetview?size=800x800&location=40.8084322,-73.9649473&heading=120&pitch=0"
Image("https://i.imgur.com/5VZ5Kft.jpg")
```

**What just happened?**

Let's break down the parameters of the API request that we just made.


    1) Root URL:             https://maps.googleapis.com/maps/api/streetview
    
    2) Question Mark:        ?
    
    3) Image Size:           size=800x800
    
    4) Ampersand:            &
    
    5) Location:             location=40.8084322,-73.9649473
    
    6) Ampersand:            &
    
    7) Heading:              heading=120
    
    8) Ampersand:            &
    
    9) Pitch:                pitch=0

**What are these parameters doing?**

Well,

    1) The root URL establishes where the API lives.
    
    2) The question mark indicates to the API that one or several parameters are coming.
    
    3) Image size is our first paremter. It indicates how big we want the image to be (maximum size is 800 x 800)
    
    4) "&" is used to chain together multiple parameters into a single API request
    
    5) Location indicates from where we want the image (Latitude, Longitude)
    
    6) Another "&" to chain together the next parameter
    
    7) Heading indicates which direction (from left to right) we want the camera to be facing
    
    8) Another "&" to chain together the next parameter
    
    9) Pitch indicates which directino (from down to up) we want the camera to be facing

Pretty cool, right?

Try changing the parameters for location, heading and pitch and seeing how it impacts the image.

**Changing pitch to 90 degrees points the camera straight up**

```python
print "https://maps.googleapis.com/maps/api/streetview?size=800x800&location=40.8084322,-73.9649473&heading=120&pitch=90"
Image("https://i.imgur.com/ugSA4dm.jpg")
```

**Changing pitch to 180 degrees points the camera all the way back**

```python
print "https://maps.googleapis.com/maps/api/streetview?size=800x800&location=40.8084322,-73.9649473&heading=120&pitch=180"
Image("https://i.imgur.com/77xdv5Q.jpg")
```

**Changing heading to 270 rotates horiztonally to the back**

```python
print "https://maps.googleapis.com/maps/api/streetview?size=800x800&location=40.8084322,-73.9649473&heading=270&pitch=0"
Image("https://i.imgur.com/URbHFiU.jpg")
```

## Scrape images from Street View along our line

We will define a function called *getStreetViewImages* and use it to cycle through every point in our line and download the corresponding image.

We will import a couple libraries to help with this.

```python
import urllib      # for URL handling
import os          # for saving images
```

We also need an *API Key* in order to make such a big request. It is common for API providers to require an API key to limit and monitor usage. This API Key is free and you may create one [here](https://developers.google.com/maps/documentation/streetview/get-api-key). Or, you can just use mine provided below.

```python
# FUNCTION INPUTS:
# 1) Our line in GeoJSON format
# 2) Desired heading (from 0 to 360)
# 3) Desired pitch (from 0 to 360)
# 4) API Key

def getStreetViewImages(geojson, heading, pitch, api_key):
    
    # Get our line coordinates
    line = geojson['features'][0]['geometry']['coordinates']
    
    # Create a list of just the latitudes
    lats = [i[1] for i in line]
    
    # Create a list of just the longitudes
    lons = [i[0] for i in line]
    
    # Keep track of which point we are on and use this to name the saved images
    count=0
    
    # For every point in our line
    for point in line:
        
        # Define the root url
        url_template = "https://maps.googleapis.com/maps/api/streetview?size={}&location={}&heading={}&pitch={}&key={}"
        
        # Define the image size as 800x800
        image_size = "800x800"
        
        # Define the location
        # Unfortunately the GeoJSON coordinates are ordered as long/lat
        # We need them to be ordered as lat/long for the Street View API
        # We also need them as a string and to remove the brackets
        # This single line takes care of those three things
        location = str([point[1],point[0]]).replace("[","").replace("]","")
        
        # Combine all of the parameters into an API request
        api_request = url_template.format(image_size, location, heading, pitch, api_key)
        
        # Print the API_request so we can see what's happening
        print api_request
        print ""
        
        # Define the image file name that we will use when saving the image
        image_file_name = str(count).zfill(5) + ".jpg"
        
        # Create an output folder to save our images
        if not os.path.exists("output"):
            os.makedirs("output")
        current_dir = os.getcwd()
        output_dir = os.path.join(current_dir,"output")
                
        # Finally! Make the API request using urllib and save the image to output folder
        urllib.urlretrieve(api_request, os.path.join(output_dir, image_file_name))
        
        # Increase our tally by one to keep track of which image we are on
        count +=1
```
